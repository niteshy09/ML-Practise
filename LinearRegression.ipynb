```python
# feature and target variable
X = df[["ENGINESIZE","CYLINDERS","FUELCONSUMPTION_COMB","FUELCONSUMPTION_COMB_MPG"]]

y = df["CO2EMISSIONS"]
```

```python
# applying standard scaler
from sklearn.preprocessing import StandardScaler

scalar = StandardScaler()
X_std = scalar.fit_transform(X)

y_std = scalar.fit_transform(y.values.reshape(-1,1))
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X_std,y_std,test_size=0.3,random_state=300)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(X_train,y_train)

y_pred = model.predict(X_test)
```

```python
error = y_test-y_pred
error
```

```python
from sklearn.metrics import mean_squared_error

from math import sqrt

sqrt(mean_squared_error(y_test,y_pred))
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```
